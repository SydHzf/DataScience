# 8.

```python
netflix['Month'] = netflix['date_added'].dt.month
netflix.sort_values(by = 'Month',inplace = True)
netflix
```

```python
hx = netflix['Month'].value_counts()
hx
```

```python
x = np.arange(0,12)
```

```python
sns.set_style('darkgrid')
plt.bar(x,hx,color = [(102/255, 1, 51/255),(1, 1, 0)] , edgecolor = 'Black')
plt.xlabel("Months")
plt.ylabel("Counts")
plt.title("Movie Release Month")
plt.xticks(x,hx.index)
plt.show()
```
