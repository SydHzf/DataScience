# 2. 

```python
m = netflix['type'].value_counts()
```

```python
x = np.arange(0,2)
x
```

```python
plt.figure(figsize = (5,5))
plt.bar(x,m,width=.5,hatch = "*+",color = ["Red","Orange"] ,edgecolor = "Black")
plt.xticks(x,m.index)
plt.xlabel("Type")
plt.ylabel("Counts")
plt.title("Content Type")
plt.grid()
plt.legend()
plt.show()
```

# 3.

```python
height = netflix["country"].value_counts().head(10)
```

```python
 x = np.arange(0,10)
plt.bar(x, height,width=0.5,color = ['Pink',"Indigo","Violet","Purple"])
plt.xticks(x,height.index,rotation = 90)
plt.xlabel("Country")
plt.ylabel("Counts")
plt.title("Top Content Creators")
plt.show()
```

# 5. and 6.



```python
movie = netflix.loc[netflix['type'] == 'Movie']
movie
tv_shows = netflix.loc[netflix['type'] == 'TV Show']
tv_shows
```

```python
x = np.arange(0,20,2)
hx = movie['director'].value_counts().sort_values(ascending=False).head(10)
hx
```

```python
y = np.arange(1,21,2)
hy = tv_shows['director'].value_counts().sort_values(ascending=False).head(10)
hy
```

```python
# MOVIES
plt.bar(x = x , height = hx ,
        width= 0.7 , 
        label = "Movies",
        color = "purple" ,
        hatch = '\\',
        edgecolor = 'pink')
plt.xticks(x,hx.index,rotation = 90)
plt.legend()
plt.show()
# TV SHOWS
plt.bar(x = y ,
        height= hy ,
        width=0.7 ,
        label = "TV Shows ",
        color  = "pink",
        hatch = '/', 
        edgecolor = "purple")
plt.xticks(y,hy.index,rotation = 90)
plt.legend()
plt.show()
```
