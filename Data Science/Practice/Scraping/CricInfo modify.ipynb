```python
import requests

from bs4 import BeautifulSoup

base_site = 'https://www.espncricinfo.com/rankings/content/page/211271.html'

response = requests.get(base_site)
response
```

```python
html = response.text
soup = BeautifulSoup(html,"html.parser")
```

```python
tables = soup.find_all("table",{"class":"StoryengineTable"})
```

```python
tables
```

```python
test_rank = tables[0]
```

```python
test_rank
```

```python
tr = test_rank.find_all("tr")
```

```python
tr
```

```python
test = []
for i in tr:
    if i.find_all("th") != []:
        test.append([i.text for i in i.find_all("th")])
    elif i.find_all("td") != []:
        test.append([i.text for i in i.find_all("td")])
test

        
```

```python
heading = [i.text for i in soup.find_all("h3")]
```

```python
    import pandas as pd
```

```python
ranking_dict = {
    
}
for i in range(len(tables)):
    table = tables[i]
    tr = table.find_all('tr')
    test = []
    for j in tr:
        if j.find_all("th") != []:
            test.append([i.text for i in j.find_all("th")])
        elif j.find_all("td") != []:
            test.append([i.text for i in j.find_all("td")])
        
    ranking = pd.DataFrame(test[1:],columns = test[0])    
    ranking_dict[heading[i]] = ranking
```

```python
ranking_dict['ICC Twenty20 Rankings']
```

```python
writer = pd.ExcelWriter("cricology.xlsx", engine = "xlsxwriter")

for k,v in ranking_dict.items():
    v.to_excel(writer,sheet_name = k,index = False)
writer.save()
```
