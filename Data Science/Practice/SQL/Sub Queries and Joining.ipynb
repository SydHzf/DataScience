# Joining

```python
q("""select f.name Country_name,
c.name Country_name,
c.population City_population,
f.population Country_population,
c.capital Capital
from facts as f
inner join cities as c
on
c.facts_id = f.id
""")
```

```python
q("""select f.name Country_name,
c.name Country_name,
c.population City_population,
f.population Country_population,
c.capital Capital
from facts as f
left join cities as c
on
c.facts_id = f.id
""")
```

```python
q("""select f.name Country_name,
c.name City_name,
c.population City_population,
f.population Country_population,
c.capital Capital
from facts as f
left join cities as c
on
c.facts_id = f.id
where City_name is Null
""")
```
