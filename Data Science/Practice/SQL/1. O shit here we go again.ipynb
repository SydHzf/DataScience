```python
def q(q):
    return pd.read_sql_query(q,conn)
```

```python
q("select Major from recent_grads")
```

```python
q("select Major,ShareWomen from recent_grads where ShareWomen > 0.5 and ShareWomen < 0.8 limit 10 ")
```

```python
q("""select Major,ShareWomen,Major_category 
from
    recent_grads 
where 
    ShareWomen >= 0.5 
or
    (Major_category = 'Engineering' 
and 
    ShareWomen <= 0.5)
order by
    ShareWomen""")
```

```python
q("""select Major,ShareWomen,Major_category 
from
    recent_grads 
where 
    ShareWomen >= 0.5 
or
    (Major_category = 'Engineering' 
and 
    ShareWomen <= 0.5)
order by
    ShareWomen desc limit 10""")
```

```python
q("""select count(ShareWomen) from recent_grads where ShareWomen > 0.5 and Major_category == "Engineering" """)
```

```python
q("""select 
count(*) 
from
recent_grads where ShareWomen > 0.5 and Major_category == "Engineering" """)
```

```python
df = q("""select count(Major) from recent_grads where Major_category == "Engineering" """)
```

```python
df.rename({"count(*)":"Total Engineering Recordes"},axis = 1)
```

```python
q("""select count(Major) as 'Total Engineering Recordes'  from recent_grads where Major_category == "Engineering" """)
```

### Group by and Having

```python
q("""select Major_category,avg(ShareWomen)
as
    Avg_ShareWomen
from
    recent_grads 
group by
    Major_category
having  
    Avg_ShareWomen > 0.5
""")# 'having' use to do things on group by 
```

```python
q("""select cast(Women as float) from recent_grads""")
```

```python
q("""select Major,cast(Women as float)/total*100 as Percentage_of_Women from recent_grads""")
```
