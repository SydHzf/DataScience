```python
def read_file(file):
    from csv import reader
    return list(reader(open(file,encoding="utf-8")))
```

```python
Art = read_file("artworks.csv")
```

```python
header = Art[0]
```

```python
data = Art[1:]
```

```python
fre = {}
```

```python
for i in data:
    artist = i[1]
    nationality = i[2]
    for j in data:
        if j[1] == artist:
            fre[artist] = [].append(j[0]),nationality    
```
