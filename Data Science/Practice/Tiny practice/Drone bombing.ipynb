```python
def read_file(file):
    from csv import reader
    return list(reader(open(file)))
```

```python
dronebombing = read_file("datasets/PakistanDroneAttacksWithTemp Ver 10 (October 19, 2017).csv")
header = dronebombing[0]
dronebombing = dronebombing[1:]
dronebombing[0
            ]
```

```python
fre = {}
def fre_table(data , col ,):
    
    for i in data:
        locations = i[col]
        if locations in fre:
            fre[locations]+=1
        else:
            fre[locations] = 1
    return fre
```

```python
# number of bomb on each location
fre_table(dronebombing,3)
```
