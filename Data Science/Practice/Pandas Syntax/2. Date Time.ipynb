```python
from csv import reader
dataset = list(reader(open("potus_visitors_2015.csv", encoding = "utf-8")))
dataset[:5]
```

```python
header = dataset[0]
dataset = dataset[1:]
```

```python
pattern = "%m/%d/%Y %H:%M"
```

```python
for i in dataset:
    i[2] = dt.datetime.strptime(i[2], pattern)
    print(i[2])
```

```python
for i in dataset:
    i[3] = dt.datetime.strptime(i[3], pattern)
```

```python
dataset[:5]
```

```python
print(dataset[0][2])
```

```python
meeting_month = {} #{2:2}

for i in dataset:
    month = i[2].month # 2
    if month in meeting_month:
        meeting_month[month] += 1
    else:
        meeting_month[month] = 1

meeting_month
```

```python
type(dataset[0][2].strftime("%d/%B-%Y"))
```

```python
dataset[0][2].strftime("%B")
```

```python
meeting_month = {} #{2:2}

for i in dataset:
    month = i[2].strftime("%b") # 2
    if month in meeting_month:
        meeting_month[month] += 1
    else:
        meeting_month[month] = 1

meeting_month
```
