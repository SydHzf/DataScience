```python
data = pd.DataFrame()
```

```python
 for i in y_axis:
#        data["Fields"] = i
        data["male"] = 100 - dataset[i]
        data["female"] = dataset[i]
        data["year"] = dataset["Year"]
        data.head().plot.barh(x = "year",
             y = ["male","female"],
             figsize = (16,6), 
             grid = False , colormap = "viridis",
             yticks = range(0,101,10),title = i)
```
